```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2  # 20% for validation
)
```

```python
train_generator = train_datagen.flow_from_directory(
    'asl_alphabet_train/asl_alphabet_train',
    target_size=(200, 200),
    batch_size=64,
    class_mode='categorical',
    subset='training',
    shuffle=True
)
```

```python
val_generator = train_datagen.flow_from_directory(
    'asl_alphabet_train/asl_alphabet_train',
    target_size=(200, 200),
    batch_size=64,
    class_mode='categorical',
    subset='validation',
    shuffle=False
)
```

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam
```

```python
# ================================
# 2. Load Base Model
# ================================
base_model = MobileNetV2(
    input_shape=(200, 200, 3),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False  # Freeze base initially
```

```python
# ================================
# 3. Build Full Model
# ================================
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
predictions = Dense(29, activation='softmax')(x)
```

```python
model = Model(inputs=base_model.input, outputs=predictions)
```

```python
# ================================
# 4. Compile & Train (Initial Phase)
# ================================
model.compile(optimizer=Adam(learning_rate=1e-3),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python

from tensorflow.keras.callbacks import EarlyStopping

print("\nüîß Initial training (frozen base)...")
model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=5,
    callbacks=[EarlyStopping(monitor='val_loss', patience=2, restore_best_weights=True)]
)
```

```python
# ================================
# 5. Fine-Tuning
# ================================
print("\nüîì Unfreezing base model for fine-tuning...")
base_model.trainable = True
```

```python
# Freeze the first 100 layers (optional, based on GPU/memory)
for layer in base_model.layers[:100]:
    layer.trainable = False
```

```python
# Recompile with a lower learning rate
model.compile(optimizer=Adam(learning_rate=1e-5),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
print("\nüéØ Fine-tuning (trainable base)...")
model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=5,
    callbacks=[EarlyStopping(monitor='val_loss', patience=2, restore_best_weights=True)]
)
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image

# Load and preprocess image
test_image = image.load_img('Nothing_test.jpg', target_size=(200, 200))  
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis=0)
test_image = test_image / 255.0  # Normalize just like training data

# Predict
result = model.predict(test_image)  
```

```python

class_indices = train_generator.class_indices

# Invert the dictionary to get index ‚Üí label mapping
labels = {v: k for k, v in class_indices.items()}

# Get predicted class index
predicted_class_index = np.argmax(result)

# Map index to class label
predicted_label = labels[predicted_class_index]

print("Predicted class:", predicted_label)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
# Get true labels
true_labels = val_generator.classes

# Get predicted labels
pred_probs = model.predict(val_generator)
pred_labels = np.argmax(pred_probs, axis=1)
```

```python
# Mapping of class indices
class_names = list(val_generator.class_indices.keys())
```

```python
# Compute the confusion matrix
cm = confusion_matrix(true_labels, pred_labels)

# Plot
plt.figure(figsize=(16, 14))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)
disp.plot(cmap=plt.cm.Blues, xticks_rotation=90)
plt.title("ASL Confusion Matrix")
plt.show()
```
